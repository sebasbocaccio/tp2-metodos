### Cargo los datos

```python
%load_ext autoreload
%autoreload 2
import pandas as pd

df_total = pd.read_csv("../data/train.csv")
```

### Tomo el 80% como conjunto de entrenamiento

```python
limite = int(len(df_total)*0.8)
df_train = df_total[:limite]
```

```python
df_train
```

### Me fijo las cantidades de cada clase en el conjunto de entrenamiento

```python
cantidades_ent = []

for i in range(10):
    cant = len(df_train[df_train['label']==i])
    cantidades_ent.append(cant)
```

```python
cantidades_ent
```

### Hago lo mismo con el conjunto de validación

```python
df_val = df_total[limite:]

cantidades_val = []

for i in range(10):
    cant = len(df_val[df_val['label']==i])
    cantidades_val.append(cant)
```

```python
cantidades_val
```

### Aplico kNN

```python
import metnum

clf = metnum.KNNClassifier(len(df_train))
```

```python
# Uso values para mandar todo a arrays de numpy
X = df_total[df_total.columns[1:]].values
y = df_total["label"].values.reshape(-1, 1)

X.shape, y.shape
```

```python
pca = metnum.PCA(15)
X = pca.transform(X)
```

```python
X_train, y_train = X[:limite], y[:limite]
X_val, y_val = X[limite:], y[limite:]

assert len(X_train) == len(y_train)
assert len(X_val) == len(y_val)

print(f"Ahora tengo {len(X_train)} instancias de entrenamiento y {len(X_val)} de validación")
```

```python
clf.fit(X_train, y_train)
```

```python
%%time
from sklearn.metrics import accuracy_score
y_pred = clf.predict(X_val)

acc = accuracy_score(y_val, y_pred)
print("Accuracy: {}".format(acc))
```
