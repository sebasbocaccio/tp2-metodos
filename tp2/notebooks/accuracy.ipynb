# Accuracy 

## KNN + PCA

```python
import csv

import pandas as pd
with open('../resultados/knn_pca_acuracy_re_do.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=(','), quotechar='|')

    for row in spamreader:
        raw_data = row
    
    # Me quedo la data sucia con "[ y ]" en los extremos, los quiero sacar.
    data = [] 
    cols = ['alpha','k','accuracy']
    result = 0
    for i in range(0,len(raw_data),3):
        data_row = [ ]
        alpha = raw_data[i][2:]
        k = raw_data[i+1][1:]
        accuracy = raw_data[i+2][1:-2]
        data_row.append(int(alpha))
        data_row.append(int(k))
        data_row.append(float(accuracy))
        data.append(data_row)
    df=pd.DataFrame(data,columns=['alpha','k','accuracy'])
    df = df.sort_values('accuracy',ascending = False)
    print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Me quedo solo con algunos alphas
plt.figure(figsize=(10,5)) 
df = df.loc[df['alpha'].isin([1,4,7,13,16,22,25])]
fig =sns.pointplot('k', 'accuracy', data=df, hue='alpha',drawstyle='default')
fig.set_xticklabels(fig.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.savefig("knn_pca_accuracy.png")
plt.show()

```
