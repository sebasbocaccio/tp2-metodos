# Precision

# Knn solo

```python
import csv

import pandas as pd
with open('../resultados/knn_solo_precision.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=(','), quotechar='|')
    numbers = next(spamreader)
    data = []
    for row in spamreader:
        data_row = [float(i) for i in row]
        data.append(data_row)
print(data)
df = pd.DataFrame(data, columns =numbers)
print(df)
```

```python


import seaborn as sns
import matplotlib.pyplot as plt
 
# image = sns.heatmap(df_1, annot=False,cmap="RdBu")
# image.set(xlabel='Digito', ylabel='K',title= 'Precision de cada digito segun el k para alpha = 1')
# fig = image.get_figure()
# fig.savefig('../resultados/knn_pca_presicion_1', dpi =500)

df_alpha1 = df.loc[df['alpha'] == 0]
dfm = df_alpha1.melt('k', var_name='dígito', value_name='precisión')
dfm = dfm[dfm['dígito'] != 'alpha']
display(dfm)
plt.figure(figsize=(9,5)) 
fig = sns.pointplot(x="k", y="precisión", hue='dígito', data=dfm, kind='point')
fig.set_xticklabels(fig.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.savefig('../resultados/knn_presicion', dpi =500)
```
